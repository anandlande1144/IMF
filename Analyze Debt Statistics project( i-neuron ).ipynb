```python
total_debt = df.loc['1970':'2030'].sum(axis =1)
df["Total Debt"]=total_debt
print(total_debt)
```

```python
max_debt_country = df.loc[df["Total Debt"].idxmax(),'Country Name']
print(max_debt_country)
```

```python
max_debt_country = df.loc[df['Total Debt'].idxmax(),"Country Name"]
max_debt_amount = df['Total Debt'].max()
print(max_debt_country, max_debt_amount)
```

```python
avg_debt_across_indi = df.groupby("Series Name")["Total Debt"].mean()
print(avg_debt_across_indi)
```

```python
#finding the most common debt indicator
most_common_indicators =df["Series Name"].mode()
most_common_indicator = most_common_indicators.max()
print(f"Most Common Indicators are\n {most_common_indicators}\n \nAnd among them is::\b {most_common_indicator}")
```

```python
#finding total debt by each countr
total_debt_countries_yearwise = df.groupby("Country Name")[df.columns[6:]].sum()
total_debt_countries = total_debt_countries_yearwise.sum(axis = 1)
print(total_debt_countries_yearwise, total_debt_countries)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
total_debt_year = df.iloc[:, 6:-1].sum()
print(total_debt_year)
```

```python
bins=[1970,1975,1980,1985,1990,1995,2000,2005,2010,2015,2020,2025,2030]
```

```python
total_debt_year.index = total_debt_year.index.astype(int)
```

```python
intervals =pd.cut(total_debt_year.index, bins= bins ,right =False)
print(intervals)
```

```python
grouped_data = total_debt_year.groupby(intervals).sum()
print(grouped_data)
```

```python
grouped_data.plot(kind = 'bar')
```
